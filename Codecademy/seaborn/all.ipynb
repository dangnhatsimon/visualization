```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
fires = pd.read_csv('fires.csv')
waste = pd.read_csv('waste.csv')
plants = pd.read_csv('plants.csv')
```

```python
%matplotlib inline
plt.rcParams['figure.dpi']=72
```

## Bar chart
Make a bar chart from the `waste` dataset that shows average `income` for each level of `gdp`. Make sure the bars run horizontally and remove the error bars.

```python
sns.barplot(data=waste,x='gdp',y='income',estimator=np.mean,ci=None)
plt.show()
```

## Scatter plot
Now make a scatter plot that compares `plastic` versus `glass` composition in the `waste` dataset.

```python
sns.scatterplot(data=waste,x='plastic',y='glass')
plt.show()
```

## Line chart
Create a line chart from the `fires` dataset that displays average `firespots` by `year`. Remove the shading around the line for the confidence interval.

```python
sns.lineplot(data=fires,x='year',y='firespots', ci=None)
plt.show()
```

## Histogram
Create a histogram that displays the distribution of `Leaf_length` from the `plants` dataset.

```python
sns.histplot(data=plants,x='Leaf_length',bins=20)
plt.show()
```

## KDE plot
Create a KDE plot from the `plants` dataset that shows the smoothed distribution of `Leaf_width`. Add a parameter that will add shading under the curve.

```python
sns.kdeplot(data=plants,x='Leaf_length',fill=True)
plt.show()
```

## Box plot
Write code to make a box plot that shows distributions of `Plant_height` across different `PH` levels in the `plants` dataset.

```python
sns.boxplot(data=plants,x='PH',y='Plant_height')
plt.show()
```
